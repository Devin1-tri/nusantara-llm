# 🇮🇩 Nusantara LLM — Dataset Exploration

Exploring the Nusantara Corpus for fine-tuning Llama 3.1 70B.

This notebook analyzes language distribution, token statistics, and quality metrics across the training data.

```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_from_disk
from transformers import AutoTokenizer

sns.set_theme(style='darkgrid')
plt.rcParams['figure.dpi'] = 120

print('Libraries loaded.')
```

```python
# Load dataset
dataset_path = '../data/processed'
dataset = load_from_disk(dataset_path)

for split in dataset:
    print(f'{split}: {len(dataset[split]):,} samples')
```

```python
# Tokenization statistics
tokenizer = AutoTokenizer.from_pretrained('meta-llama/Llama-3.1-70B')

lengths = [len(ids) for ids in dataset['train']['input_ids']]
print(f'Mean length: {np.mean(lengths):.1f}')
print(f'Median length: {np.median(lengths):.1f}')
print(f'Max length: {np.max(lengths)}')
print(f'Total tokens: {np.sum(lengths):,}')

plt.figure(figsize=(10, 4))
plt.hist(lengths, bins=100, alpha=0.7)
plt.xlabel('Sequence Length (tokens)')
plt.ylabel('Count')
plt.title('Training Data: Sequence Length Distribution')
plt.show()
```

```python
# Language distribution analysis
print('Dataset statistics saved to: dataset_stats.json')
with open(f'{dataset_path}/dataset_stats.json') as f:
    stats = json.load(f)
print(json.dumps(stats, indent=2))
```
